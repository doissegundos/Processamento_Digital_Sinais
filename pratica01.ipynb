```
import numpy as np
import matplotlib.pyplot as plt
```

1-) Gere um sinal que é a soma de três funções seno com freqüências angulares
discretas iguais a 0,1π; 0,5π e 0,75π rad/s; e amplitudes iguais a 1; 1.5 e 0.5;
respectivamente. Estas seqüências devem possuir 201 pontos (0 ≤ n ≤ 200). Visualize
este sinal no tempo [sin, :, figure, plot].


```
s =  np.arange(0,201,1)
sinal1 = 1 * np.sin((0.1*np.pi)*s)
sinal2 = 1.5 * np.sin((0.5*np.pi)*s)
sinal3 = 0.5 * np.sin((0.75*np.pi)*s)
sinal_soma = sinal1 + sinal2 + sinal3
```

```
plt.plot(s,sinal_soma)
plt.show()
```

2-) Gere a saída de um sistema de média móvel com M1=0 e M2=3, sendo a entrada
dada pelo sinal gerado no item 1 [mean]. Desenho o gráfico deste sinal na mesma figura
do item 1 [hold].

```
y = []
for i in range(196):
  y.append((sinal_soma[i]+sinal_soma[i+1]+sinal_soma[i+2]+sinal_soma[i+3])/4)
```

```
plt.plot(s,sinal_soma)
plt.plot(y)
plt.show()
```

3-) Gere um sinal que é a soma de três funções degrau unitário deslocadas no tempo
de 0, 20 e 40 amostras, respectivamente. Cada uma destas sequências deve possuir
N=201 pontos, com 0≤ n ≤ 200. Gere o gráfico deste sinal [zeros, ones].

```
s =  np.arange(0,201,1)
```

```
u1 = np.ones(201)

u2 = list(np.zeros(20))
u2.extend(np.zeros(181))

u3 = list(np.zeros(40))
u3.extend(np.zeros(161))
```

```
x = u1 + u2 + u3
```

```
plt.plot(s,x)
plt.show()
```

4-) Gere a saída de um sistema de média móvel com M1=0 e M2 = 10, sendo a entrada
dada pelo sinal gerado no item 3 [mean]. Desenho o gráfico deste sinal na mesma figura
do item 3. 

```
y = []
aux = 0
for i in range(190):
  for j in range(11):
    aux = aux + x[i+j]
  y.append(aux/11)
  aux = 0
```

```
plt.plot(s,x)
plt.plot(y)
plt.show()
```

5-) Gere o seguinte sinal sinc: x[n] = sen(0,3 π n)/ 0,3 π n, com -100 ≤ n ≤ 100. Gere o
gráfico deste sinal [sin].

```
s =  np.arange(-100,101,1)
sinal = np.sin((0.3*np.pi)*s) / (0.3*np.pi)*s
```

```
plt.plot(s,sinal)
plt.show()
```

6-) Gere a saída de um sistema acumulador, sendo a entrada dada pelo sinal gerado no
item 5 [sum]. Desenho o gráfico deste sinal na mesma figura do item 5. 

```
y = []
y.append(sinal[0])
for i in range(1,201,1):
  y.append(sinal[i] + y[i-1])
```

```
plt.plot(s,sinal)
plt.plot(y)
plt.show()
```

7-) Gere o seguinte sinal exponencial: x[n] = (1/3)^n u[n], em que u[n] é a função degrau
unitário, com 0≤ n ≤ 100. Gere o gráfico deste sinal [^, figure, plot ou stem].


```
s =  np.arange(0,101,1)
u = np.ones(101)
```

```
sinal = ((1/3)**s) * u
```

```
plt.plot(s,sinal)
plt.show()
```

8-) Gere a saída de um sistema acumulador, sendo a entrada dada pelo sinal gerado no
item 7 [sum]. Desenho o gráfico deste sinal na mesma figura do item 7.

```
y = []
y.append(sinal[0])
for i in range(1,101,1):
  y.append(sinal[i] + y[i-1])
```
